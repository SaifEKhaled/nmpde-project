```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../data/convergence_results.csv')
r = df['Refinement'].values
errors = df['Error'].values

h = 30 / (2**r)
```

```python
plt.figure(figsize=(8, 6))
plt.loglog(h, errors, 's-', markersize=10, linewidth=3, label='Our Matrix-Free RK4 ($Q_4$)')
```

```python
ref_line = errors[0] * (h / h[0])**5
plt.loglog(h, ref_line, 'k--', alpha=0.6, label='Theoretical $O(h^5)$')

plt.xlabel('Mesh Size ($h$)', fontsize=12, fontweight='bold')
plt.ylabel('$L_2$ Error Norm', fontsize=12, fontweight='bold')
plt.title('Spatial Convergence Analysis: Wave Equation', fontsize=14, fontweight='bold')
plt.grid(True, which="both", ls="-", alpha=0.5)
plt.legend()
```

```python
slopes = np.diff(np.log(errors)) / np.diff(np.log(h))
print(f"Observed Convergence Rate: {slopes.mean():.2f}")

plt.savefig('../results/convergence_graph.png', dpi=300)
plt.show()
```
